#### STAGE 2 - Finding "h" < n  least outlying data points using the Stahel-Donoho estimator
like in the paper h = 0.7*n = 126 

**To find the least outlying data points we use the projection pursuit principle : Multivariate Outlier should be outlier in atleast one direction but not necessarily the direction of the coordinate axes**


**So we consider a set B of the directions to project along and then we project all the datapoints onto a particular vector $v \in B$  and now for every datapoint we use the Stahel-Donoho measure to  quantify how "outlying" the data point is in direction v . We find this measure of "outlyingness"  $\forall v \in B$ and the measure of "outlyingness" for a data point  $x_{i}$ is the maximum of this value over all directions in B**


**B consists of 250 vectors which are linear combinations of any two data points**

```python
### function to get the set B -- set of random directions to project 
## 250 Directions each of them are obatained from 2 randomnly sampled points from the data set 
def get_Directions(X,seed):
    mat = np.zeros((250,X.shape[1]))
    for i in range(250):
        np.random.seed(seed+i)
        random_points = np.random.choice(X.shape[0], 2, replace=False)
        mat[i,:] = X.iloc[random_points[0],:] - X.iloc[random_points[1],:]
    return pd.DataFrame(mat)
B = get_Directions(new_data,0)
print(B)
      
```

```python
from sklearn.covariance import MinCovDet
## now we need to project all the datapoints and find the Stahel-Donoho measure for each observation
def get_OutlierScores(X,h = 0.7):
    B = get_Directions(X,0)
    temp = np.sqrt(np.square(B).sum(axis=1)) ## finding norm of each row in B ( IE direction)
    ## since we want non zero B we must remove those vectors with zero norm 
    B_reduced = B[temp > 1e-12]
    temp2 = temp[temp > 1e-12]
    projections = X @ np.transpose(np.diag(1/temp2) @ B_reduced)
    Outlyingness = np.zeros((X.shape[0],len(temp2)))## collect outlyingness of each point in the non zero directions
    for i in range(len(temp2)):
        x = projections.iloc[:,i]
        cov = MinCovDet(random_state=0,support_fraction= h).fit(np.array(x).reshape(-1,1))
        Outlyingness[:,i] = abs(projections.iloc[:,i].values-cov.location_)/np.sqrt(cov.covariance_)
    Outlying_scores = np.amax(Outlyingness,axis=1)
    return Outlying_scores
```
